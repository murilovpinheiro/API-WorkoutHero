```python
import re

def funcao(linha):
    try:
        # linha2 = '_'.join(linha.replace('(male)', '').replace('(female)', '').strip().split(' '))
        # return linha2
        linha2 = re.sub(r'\([^)]*\)', '', linha).strip()
        return '_'.join(linha2.split(' '))
    except:
        return linha
    
df2['name'] = df2['name'].apply(funcao)
df2['name']
```

```python
import requests

def salvar_imagem(url, caminho_destino, nome_arquivo):
    nome_completo = caminho_destino + '/' + nome_arquivo.replace('/', '-')
    
    var = requests.get(url)
    with open(nome_completo, 'wb') as f:
        f.write(var.content)

    print("Imagem salva com sucesso!")
```
